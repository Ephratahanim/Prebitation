```python
%load_ext autoreload
%autoreload 2

# First step, import libraries.
import numpy as np 
import pandas as pd 
import seaborn as sb
from matplotlib import pyplot as plt
from utils import DatasetV1
%matplotlib inline
```

```python
DATA_PATH = "../data/bitcoin-historical-data/coinbaseUSD_1-min_data_2014-12-01_to_2017-10-20.csv.csv"
PRED_DAYS = 10  # num prediction days for the trained models
TIME_GRAN = 1  # time granularity in minutes
STEP_SIZES = [180, 360, 720]
TRAIN_START = 1200000
```

### Prepare Data

```python
STEP_SIZE = STEP_SIZES[0]  # 180 for testing
num_test_steps = ((24*60) / TIME_GRAN) * PRED_DAYS 
dataset = DatasetV1(DATA_PATH,
                    train_start=1200000,
                    test_start=num_test_steps,
                    time_gran=TIME_GRAN)

# split data
dataset.create_train_data_v4(["Close"],
                              "Close",
                              time_steps=STEP_SIZE,
                              preprocess=False)
```

### Clustering Data

```python
from sklearn.cluster import MiniBatchKMeans, KMeans
# kmeans = MiniBatchKMeans(n_clusters=100, init='k-means++', max_iter=100,
#                          batch_size=128, verbose=0, compute_labels=True,
#                          random_state=None, tol=0.0, max_no_improvement=10,
#                          init_size=None, n_init=3, reassignment_ratio=0.01)
kmeans = KMeans(n_clusters=100, init='k-means++', n_init=10, max_iter=300, tol=1e-6,
                precompute_distances='auto', verbose=12, random_state=None,
                copy_x=True, n_jobs=-1, algorithm='auto')
```

```python
kmeans.fit(dataset.X_train[])
```

```python
# assing items to clusters
c_idxs = kmeans.predict(dataset.X_train)
# and get cluster centroids
cs = kmeans.cluster_centers_
```

```python
# compute distance item to all centroids
dists = kmeans.transform(dataset.X_train)
```

```python
# Find cluster populations
item_counts = np.bincount(c_idxs)
# sort clusters by population 
c_idx_sort = np.argsort(-np.bincount(c_idxs))
counts_sort = -np.sort(-np.bincount(item_counts))
```

```python
dataset.y_train[c_idxs==0]
```

```python
# find weightes average price per cluster
c_stats = []
for c_idx in range(cs.shape[0]):
    vals = dataset.y_train[c_idxs==c_idx]
    c_stats.append(vals)
```

```python
plt.figure(figsize=[20,20])
for i in range(20):
    plt.subplot(10,4,i*2+1)
    idx = c_sort[i]
    plt.plot(cs[idx])
    plt.subplot(10,4,i*2+2)
    plt.plot(np.sort(c_stats[idx]))
    title = "{} {} {}".format(c_stats[idx].mean(), c_stats[idx].max(), c_stats[idx].min())
    plt.title(title)
```
